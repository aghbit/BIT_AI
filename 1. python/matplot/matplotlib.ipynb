```python
import numpy as np
import matplotlib.pyplot as plt
```

# [Plotting](https://matplotlib.org/tutorials/introductory/pyplot.html)

Funkcja `plot` przyjmuje współrzędne punktów i łonczy ich prostą liniją, ma opcjonalne parametry: `color`(default: niebieski), `markers`(dla zaznaczenia punktów zadanych paramentrem X lub `markevery`)

```python
x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)

plt.plot(x,y, marker='o', markevery=5)# markuj co piąty punkt # lub markevery=[1, 5] wtedy będzie markowany tylko pierwszy i piąty punkt
plt.show()
```

Parametr `xlabel` ustawia napis na osi OX <br/> `ylabel` ustawia napis na osi OY <br/> `title` ustawia nadgłówek nad wykresem <br/> `legend` ustawia legendę wykresu(są 2 możliwości ustawienia nazw) <br/> `label` parametr który markuje wykres 

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Sinus i cosinus")
plt.legend(["sin", "cos"]) # pierwsza metoda, załeży od kolejności podanych argumentów plt.legend(["cos", "sin"])
plt.show()
```

```python
plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Sinus i cosinus")
plt.legend() # metoda nie zależy od kolejności argómentów(nie ma parametrów, bo pobiera ich z parametru `label`)
plt.show()
```

# Subploting 

Na jednym polu dla rysowania można narysować kilka niezależnych wykresów.<br/> Najprościej używać dla tego metodę `subplot` która ustawia wykresy w siatkę(grid)<br/><br/> **Parametry:** <br/> Pierwszy parametr przyjmuje `ilość wierszy`<br/> Drugi `ilość kolumn` <br/> Trzeci `numer wykresu` na którym będziemy rysować (można wyliczyć wzorem `numer_wiersza * ilość_kolumn + numer kolumny` )

```python
plt.subplot(2,1,1)
plt.plot(x, y_sin)
plt.title("Sinus")

plt.subplot(2,1,2)
plt.plot(x, y_cos)
plt.title("Cosinus")

plt.tight_layout() # używamy kiedy 2 wykresy pokrywają się z sobą (sprobój )

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y_sin)
plt.title("Sinus")

plt.subplot(1,2,2)
plt.plot(x, y_cos)
plt.title("Cosinus")

plt.tight_layout()
plt.show()
```

Używanie parametru `color`

```python
plt.subplot(2,1,1)
plt.plot(x, y_sin, color='#00FF00', lw=5, label="Sinus") # lw(line width)(grubość linii)
plt.title("Title")
plt.legend()

plt.subplot(2,1,2)
plt.plot(x,y_cos, color='g', label="Cosinus")
plt.axis('off')
plt.legend()

plt.show()
```

```python
x = np.arange(7)
y = x
plt.plot(x, y, color='#00FF00', marker='o')
plt.show()
```

# Scattering

Funkcja `scatter` przyjmuje jako argument `współrzędne punktów`, `kolor`(default: niebieski), `typ markeru`(default: kółko) 

```python
X = np.array([1, 2, 3, 4, 5, 6, 7])
Y = np.array([4.1, 6.7, 10.8, 14.3, 15.5, 20.0, 21.37])

plt.scatter(X, Y)
plt.show()
```

Na jednej płaszczyźnie mogą być kilka wykresów rzóżnych typów: 

```python
w2, w1, w0 = np.polyfit(X, Y, deg=2) # dopasowuje wielomian deg-stopnia do podanych punktów 
y_ap = w0 + w1 * X + w2 * (X ** 2)
plt.scatter(X,Y, color='#AFFF00')
plt.plot(X, y_ap, color='red')
plt.show()
```

```python
x = np.arange(1, 8, 0.2)
y_log = np.log(x) # naturalny logarytm
y_log10 = np.log10(x) # logarytm o podstawie 10

plt.scatter(x, y_log, marker='3', color='g', label='log')
plt.scatter(x, y_log10, marker='_', color='b', label='log10')
plt.legend()
plt.show()
```

[Spis wszystkich markerów](https://matplotlib.org/api/markers_api.html)

```python
markers=['.',',','o','v','^','<','>','1','2','3','4','8','s','p','P','*','h','H','+','x','X','D','d','|','_']
descriptions=['point', 'pixel', 'circle', 'triangle_down', 'triangle_up','triangle_left', 'triangle_right', 'tri_down', 'tri_up', 'tri_left','tri_right', 'octagon', 'square', 'pentagon', 'plus (filled)','star', 'hexagon1', 'hexagon2', 'plus', 'x', 'x (filled)','diamond', 'thin_diamond', 'vline', 'hline']

n = len(markers)
r = n//5
c = n//5

ii = np.arange(r)
jj = np.arange(c)

for i in ii:
    for j in jj:
        plt.text(i-0.15, j+0.15, s='\'' + markers[i*r+j] + '\'' + ' : ' + descriptions[i*r+j], fontsize=8)
        plt.scatter(i, j, marker=markers[i*r+j])
        
plt.axis('off')
plt.tight_layout()
plt.show()
    
```

# [Histograms](https://matplotlib.org/gallery/statistics/hist.html)

`hist` przyjmuje wektor `x`, parametr `bins` specyfikuje ilość "schodków" 

```python
x = np.random.randn(100000)
y = np.random.randn(100000)

bins = 30

plt.hist(x, bins=bins)
plt.show()
```

```python
plt.hist2d(x+1,y-2) 
plt.show() # jest żółty na miejscu maksimum 
```

```python
plt.hist2d(x+1,y-2, bins=(40,80)) # jest żółty na miejscu maksimum 
plt.show()
```

# Images

```python
import matplotlib.image as mpimg

img = mpimg.imread('kotek.jpeg') # wczypuje zdjęcie jak np.array 
print(img.dtype, img.shape) # specjalny typ np.uint8 

img_tinted = img * [0.7, 1, 0.95] # zielony kanal zostaje ten sam, czerwony i niebieski przyciemnione 

print(img_tinted.dtype) # float64, czyli plt już nie jest w stanie potraktować to jak zdjęcie

plt.subplot(1,2,1)
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.imshow(img_tinted.astype(np.uint8)) # żeby zmienić typ z float64 na uint8 używamy .astype(<nazwa_typu>)

plt.show()
```
